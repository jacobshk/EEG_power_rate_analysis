```python
def calculate_frame_shifted_score(df):
    """
    Calculate score for every Score Event (i.e. when user_answer and correct_answer are not NaN)
    Score is based on positional matching:
    - Compare user_answer and correct_answer position by position (same index)
    - Find the first contiguous subsequence of length > 1 where elements match at the same positions
    - Score is the length of this subsequence (0 if no match > 1 found)
    
    Returns a df containing only rows with valid user_answer and correct_answer, plus a 'score' column
    """
    import pandas as pd
    
    # Filter to only rows where both user_answer and correct_answer are not NaN
    mask = df['user_answer'].notna() & df['correct_answer'].notna()
    result_df = df[mask].copy()
    
    # Initialize score column
    result_df['frame_shifted_score'] = 0
    
    for idx in result_df.index:
        user_ans = str(result_df.loc[idx, 'user_answer']).split('-')
        correct_ans = str(result_df.loc[idx, 'correct_answer']).split('-')
        
        # Ensure both lists have the same length (use minimum)
        min_len = min(len(user_ans), len(correct_ans))
        
        score = 0
        i = 0
        
        while i < min_len:
            # Count consecutive matches starting at position i
            match_length = 0
            j = i
            while j < min_len and user_ans[j] == correct_ans[j]:
                match_length += 1
                j += 1
            
            # If we found a match of length > 1, that's our score
            if match_length > 1:
                score = match_length
                break
            
            # Move to next position
            i += 1
        
        result_df.loc[idx, 'frame_shifted_score'] = score
    
    return result_df
```

```python
def calculate_direct_match_score(df):
    """
    Add a 'direct_match' column based on left-to-right sequential matching.
    
    Score is the number of consecutive matches starting from the leftmost position (index 0).
    Stops counting at the first mismatch.
    
    Examples:
    - user: 1-2-3-5-4-7-6-8, correct: 1-2-3-4-5-6-7-8 → direct_match = 3 (matches at positions 0,1,2)
    - user: 1-3-2-6-4-5-7-8, correct: 1-2-3-4-5-6-7-8 → direct_match = 1 (matches only at position 0)
    - user: 2-1-3-4-5-6-7-8, correct: 1-2-3-4-5-6-7-8 → direct_match = 0 (no match at position 0)
    
    Returns the df with an additional 'direct_match' column
    """
    import pandas as pd
    
    # Create a copy to avoid modifying the original
    result_df = df.copy()
    
    # Initialize direct_match column
    result_df['direct_match_score'] = 0
    
    for idx in result_df.index:
        user_ans = str(result_df.loc[idx, 'user_answer']).split('-')
        correct_ans = str(result_df.loc[idx, 'correct_answer']).split('-')
        
        # Count consecutive matches from the start
        min_len = min(len(user_ans), len(correct_ans))
        direct_match = 0
        
        for i in range(min_len):
            if user_ans[i] == correct_ans[i]:
                direct_match += 1
            else:
                # Stop at first mismatch
                break
        
        result_df.loc[idx, 'direct_match_score'] = direct_match
    
    return result_df
```
