```python
#can we see the electrode locations in 3d? https://mne.tools/stable/auto_tutorials/intro/40_sensor_locations.html#visualizing-sensors-in-3d-surface-renderings

filepath = '/scratch/jsheikh2/ted_temp/ds005505/sub-NDARGJ395FKP/eeg/sub-NDARGJ395FKP_task-seqLearning8target_eeg.set'

import mne
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load your data
raw = mne.io.read_raw_eeglab(filepath, preload=True)

# Method 1: MNE's built-in 3D sensor plot
fig = raw.plot_sensors(kind='3d', show_names=True)

# Method 2: Interactive 3D plot with matplotlib
montage = raw.get_montage()
if montage:
    pos = montage.get_positions()
    ch_pos = pos['ch_pos']
    
    # Extract coordinates
    ch_names = list(ch_pos.keys())
    coords = np.array([ch_pos[ch] for ch in ch_names])
    
    # Create 3D plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    ax.scatter(coords[:, 0], coords[:, 1], coords[:, 2], 
               c='red', marker='o', s=100)
    
    # Add labels
    for i, name in enumerate(ch_names):
        ax.text(coords[i, 0], coords[i, 1], coords[i, 2], 
                name, fontsize=8)
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('3D Electrode Positions')
    plt.show()

# Method 3: Plot on brain surface (requires FreeSurfer subject or template)
#mne.viz.plot_alignment(raw.info, trans='
```
