```python
# HELPER FUNCTIONS
import numpy as np
import mne
from scipy import signal
def calculate_event_rate_zscore_optimized(data, sfreq, band, z_threshold=2.0):
    """
    
    Calculate event rate as crossings above z-score threshold.
    
    Parameters:
    -----------
    data : array (n_channels, n_samples)
    sfreq : float, sampling frequency
    band : tuple, (low_freq, high_freq)
    z_threshold : float, z-score threshold (default: 2.0)
    
    Returns:
    --------
    event_rate : float, events per minute averaged across channels
    n_events : int, total number of events across all channels
    """
    
    duration_min = data.shape[1] / (sfreq * 60)

    # Bandpass filter
    sos = signal.butter(4, band, btype='bandpass', fs=sfreq, output='sos')
    filtered_data = signal.sosfiltfilt(sos, data, axis=-1)
    
    # Calculate envelope
    analytic_signal = signal.hilbert(filtered_data, axis=-1)
    envelope = np.abs(analytic_signal)
    
    # Z-score normalization
    z_scored = (envelope - envelope.mean(axis=1, keepdims=True)) / envelope.std(axis=1, keepdims=True)
    
    # Detect crossings
    above_threshold = z_scored > z_threshold
    crossings = np.diff(above_threshold.astype(int), axis=-1)
    n_events_per_channel = np.sum(crossings == 1, axis=-1)
    n_events_mean = np.mean(n_events_per_channel)
    
    event_rate = n_events_mean / duration_min
    return event_rate

def bandpower(psd, freqs, band):
    """Calculate mean power in a frequency band."""
    # Boolean mask for frequencies in this band
    freq_mask = (freqs >= band[0]) & (freqs <= band[1])
    
    # Mean across frequency bins, then across channels
    # Shape: (n_channels, n_freqs) -> (n_channels,) -> scalar
    band_power = psd[:, freq_mask].mean(axis=-1).mean()
    
    return band_power
```

```python
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm import tqdm
import numpy as np
import mne

def process_single_participant(row, bands):
    """Process a single participant - designed for parallel execution."""
    participant_id = row['participant_id']
    idx = row.name  # Get the original index
    baseline_filepath = f'/scratch/jsheikh2/ted_temp/ds005505/{participant_id}/eeg/{participant_id}_task-RestingState_eeg.set'
    
    result_row = {'index': idx, 'participant_id': participant_id}
    
    try:
        raw = mne.io.read_raw_eeglab(baseline_filepath, preload=True, verbose=False)
        
        # Load data ONCE
        data = raw.get_data()
        
        # Identify bad channels in one step:
        # - Zero standard deviation (flat line)
        # - All zero values
        # - Near-zero variance
        channel_stds = data.std(axis=1)
        channel_means = np.abs(data).mean(axis=1)
        
        # A channel is bad if it has zero std OR is all zeros
        bad_channel_mask = (channel_stds <= 1e-20) | (channel_means <= 1e-20)
        bad_channel_indices = np.where(bad_channel_mask)[0]
        bad_channel_names = [raw.ch_names[i] for i in bad_channel_indices]
        
        # Store the bad channels that were filtered out
        result_row['bad_channels'] = ','.join(bad_channel_names) if bad_channel_names else ''
        
        # Drop bad channels if any exist
        if len(bad_channel_indices) > 0:
            good_channel_indices = np.where(~bad_channel_mask)[0]
            
            if len(good_channel_indices) == 0:
                raise ValueError(f"All channels are bad for {participant_id}")
            
            data = data[good_channel_indices, :]
            raw = raw.copy().pick_channels([raw.ch_names[i] for i in good_channel_indices])
        
        sfreq = raw.info['sfreq']
        
        # Compute PSD once (should not have warnings now)
        psd = raw.compute_psd(verbose=False)
        freqs = psd.freqs
        psd_data = psd.data
        
        # Calculate metrics for each band
        for band_name, band_range in bands.items():
            # Power from pre-computed PSD
            result_row[f'{band_name}_baseline_power (μV²/Hz)'] = \
                bandpower(psd_data, freqs, band_range)
            # Event rate using already-loaded data
            result_row[f'{band_name}_baseline_event_rate'] = \
                calculate_event_rate_zscore_optimized(data, sfreq, band_range)
        
        return result_row
        
    except Exception as e:
        print(f"Error processing {participant_id}: {e}")
        return {'index': idx, 'participant_id': participant_id, 'error': str(e)}


def compute_power_event_rate_df(participants_df, bands, n_workers=8):
    """
    Process pool version for parallel processing.
    
    Parameters:
    -----------
    participants_df : DataFrame with 'participant_id' column
    bands : dict of band names and frequency ranges
    n_workers : int, number of workers to use (default: 8)
    """
    
    # Pre-initialize columns
    participants_df['bad_channels'] = ''  # Initialize bad channels column
    for band_name in bands.keys():
        participants_df[f'{band_name}_baseline_power'] = np.nan
        participants_df[f'{band_name}_baseline_event_rate'] = np.nan
    
    print(f"Processing {len(participants_df)} participants using {n_workers} workers...")
    
    # Convert dataframe rows to list
    rows = [row for _, row in participants_df.iterrows()]
    
    # Process pool execution with progress bar
    results = []
    with ProcessPoolExecutor(max_workers=n_workers) as executor:
        # Submit all tasks
        future_to_row = {executor.submit(process_single_participant, row, bands): row 
                         for row in rows}
        
        # Collect results with progress bar
        for future in tqdm(as_completed(future_to_row), total=len(rows)):
            try:
                result = future.result()
                results.append(result)
            except Exception as e:
                row = future_to_row[future]
                print(f"Process exception for {row['participant_id']}: {e}")
    
    # Batch update DataFrame
    for result in results:
        if result is None or 'error' in result:
            continue
            
        idx = result.pop('index')
        result.pop('participant_id')  # Remove redundant field
        
        for col, val in result.items():
            participants_df.loc[idx, col] = val
    
    return participants_df


# Usage examples:
bands = {
    'Delta': (0.5, 4),
    'Theta': (4, 12),
    'Alpha': (8, 13),
    'Beta': (13, 30),
    'Slow gamma': (25, 55),
    'Fast gamma': (60, 100),
    'SWR (human)': (80, 150)
}

display(participants_df)

participants_df = compute_power_event_rate_df(participants_df, bands, n_workers=32)
```

```python
import mne
import numpy as np
from mne.datasets import fetch_fsaverage

def map_gsn129_to_regions(raw, parc='aparc'):
    """Map GSN HydroCel 129 electrodes to brain regions."""
    
    # 1. Set up template brain
    fs_dir = fetch_fsaverage()
    subjects_dir = fs_dir.parent
    
    # 2. Apply GSN-HydroCel-129 montage
    montage = mne.channels.make_standard_montage('GSN-HydroCel-129')
    raw.set_montage(montage)
    
    # 3. Load source space and BEM
    src = mne.read_source_spaces(fs_dir / 'bem' / 'fsaverage-ico-5-src.fif')
    bem = mne.read_bem_solution(fs_dir / 'bem' / 'fsaverage-5120-5120-5120-bem-sol.fif')
    
    # 4. Create forward solution
    fwd = mne.make_forward_solution(
        raw.info, trans='fsaverage', src=src, bem=bem, eeg=True
    )
    fwd = mne.convert_forward_solution(fwd, surf_ori=True, force_fixed=True)
    
    # 5. Load parcellation labels
    labels = mne.read_labels_from_annot('fsaverage', parc, subjects_dir=subjects_dir)
    
    # 6. Compute electrode-to-region sensitivity
    leadfield = np.abs(fwd['sol']['data'])
    electrode_mapping = {}
    
    for ch_idx, ch_name in enumerate(fwd['info']['ch_names']):
        sensitivity = leadfield[ch_idx, :]
        label_scores = []
        
        for label in labels:
            verts = label.get_vertices_used()
            if len(verts) > 0:
                score = sensitivity[verts].mean()
                label_scores.append((label.name, score))
        
        electrode_mapping[ch_name] = sorted(label_scores, key=lambda x: -x[1])[:5]
    
    return electrode_mapping



path = '/scratch/jsheikh2/ted_temp/ds005505/sub-NDARGJ395FKP/eeg/sub-NDARGJ395FKP_task-seqLearning8target_eeg.set'
#https://mne.tools/stable/generated/mne.io.read_raw_eeglab.html
raw = mne.io.read_raw_eeglab(path)

map_gsn129_to_regions(raw)
```
